```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
csv_file = "MetObjects.csv"
db_file = "MetObjects.db"
table_name = "MetObjects"

df = pd.read_csv(csv_file)

conn = sqlite3.connect(db_file)

df.to_sql(table_name, conn, if_exists="replace", index=False)

conn.close()
```

Display the first 10 rows of the `MetObjects` table.

```python
conn = sqlite3.connect('MetObjects.db')
display(pd.read_sql_query('SELECT * FROM MetObjects LIMIT 10', con=conn))
conn.close()
```
